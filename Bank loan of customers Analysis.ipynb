# Year wise loan amount Stats

#### Convert the Date Column to DateTime Format

```python
# Convert the 'Loan Date' column to datetime if it's not already
Finance1["issue_d"] = pd.to_datetime(Finance1["issue_d"])
```

```python
Finance1.head()
```

```python
# Extract the year from 'issue_d'
Finance1['year'] = Finance1['issue_d'].dt.year

# Display the first few rows of the DataFrame
print(Finance1[['issue_d', 'year']].head())
```

```python
Finance1.head()
```

```python
Finance1 = Finance1.dropna(subset=['issue_d'])
```

#### Group the Data by Year and Calculate Statistics

```python
# Group by 'Year' and calculate total loan amount and average loan amount
yearly_stats = Finance1.groupby('year').agg(
    total_loan_amount=('loan_amnt', 'sum'),
    average_loan_amount=('loan_amnt', 'mean'),
    loan_count=('loan_amnt', 'count')
).reset_index()

# Display the results
print(yearly_stats)
```

#### Visualize the Results

```python
# Optionally, save the statistics to a new Excel file
yearly_stats.to_excel("yearly_loan_stats.xlsx", index=False)
```

```python
# Total Loan Amount Over the Years
plt.figure(figsize=(10,6))
plt.bar(yearly_stats['year'], yearly_stats['total_loan_amount'], color='b', label='Total Loan Amount')
plt.xlabel('Year')
plt.ylabel('Total Loan Amount')
plt.title('Year-wise Total Loan Amount')
plt.grid(True, axis='y')
plt.legend()
plt.show()
```

```python
# Average Loan Amount Over the Years

plt.figure(figsize=(10,6))
plt.bar(yearly_stats['year'], yearly_stats['average_loan_amount'], color='g', label='Average Loan Amount')
plt.xlabel('Year')
plt.ylabel('Average Loan Amount')
plt.title('Year-wise Average Loan Amount')
plt.grid(True, axis='y')
plt.legend()
plt.show()
```

# Grade & Sub_Grade wise revol_bal

```python
# Merge finance1 and finance2 on common column
merged_df = pd.merge(Finance1, Finance2, on= "id" , how = "inner" )
```

```python
# Group by grade and sub_grade and calculate sum of revol_bal
grouped_revol_balc = merged_df.groupby(["grade","sub_grade"])["revol_bal"].sum().reset_index()
```

```python
# Sort the results
grouped_revol_balc = grouped_revol_balc.sort_values(by=["grade","sub_grade"])
```

```python
print(grouped_revol_balc)
```

```python
pivot_df = grouped_revol_balc.pivot(index='sub_grade', columns='grade', values='revol_bal').fillna(0)
```

# Visualize the Results

```python
pivot_df.plot(kind='bar', figsize=(10,6))
plt.title('Revolving Balance by Sub_Grade and Grade')
plt.xlabel('Sub Grade')
plt.ylabel('Revolving Balance')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.legend(title='Grade')
plt.tight_layout()
plt.show()
```

```python
# Pivot the data for plotting
pivot_df1 = grouped.pivot_table(index=['addr_state', 'Month'], 
                                columns='loan_status', 
                                values='count', 
                                fill_value=0)

# Optional: Flatten the multi-index
pivot_df1 = pivot_df1.reset_index()
```

```python
# Filter for one state (like 'CA')
ca_df = pivot_df1[pivot_df1['addr_state'] == 'CA'].set_index('Month')

# Drop 'addr_state' column for plotting
ca_df = ca_df.drop(columns='addr_state')

# Plot stacked bar
ca_df.plot(kind='bar', stacked=True, figsize=(10,6))
plt.title('Monthly Loan Status Distribution - California')
plt.xlabel('Month')
plt.ylabel('Loan Count')
plt.xticks(rotation=45)
plt.legend(title='Loan Status')
plt.tight_layout()
plt.show()
```

### Find Top 10 States by Total Loan Count

```python
top_states = merged_df['addr_state'].value_counts().nlargest(10).index.tolist()
```

```python
# Filter Data for Only Those States
top_df = merged_df[merged_df['addr_state'].isin(top_states)]
```

```python
# Trend Line Chart (Top 10 States, Overall Loan Status)
trend_df = top_df.groupby(['Month', 'loan_status']).size().reset_index(name='count')

plt.figure(figsize=(12,6))
sns.lineplot(data=trend_df, x='Month', y='count', hue='loan_status', marker='o')
plt.title('Trend of Loan Status Over Time (Top 10 States)')
plt.xlabel('Month')
plt.ylabel('Loan Count')
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Heatmap (Loan Volume by State and Month – Top 10 Only)
heatmap_data = top_df.groupby(['addr_state', 'Month']).size().unstack(fill_value=0)

plt.figure(figsize=(14,8))
sns.heatmap(heatmap_data, cmap='YlGnBu', annot=True, fmt='d', linewidths=0.5)
plt.title('Loan Volume Heatmap (Top 10 States)')
plt.xlabel('Month')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```
